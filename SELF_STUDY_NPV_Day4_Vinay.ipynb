```python
plt.bar(stock_names , stock_values)
plt.show()
```

```python
plt.bar(stock_names , stock_values , color = ['r','g'])
plt.show()
```

```python
plt.bar(stock_names , stock_values , color = ['r','g'])
plt.bar(stock_names , emp_count , color = ['b','c'])
plt.show()
```

```python
#stacked bar chart
plt.bar(stock_names , stock_values , color = ['r','g'])
plt.bar(stock_names , emp_count , bottom = stock_values , color = ['b','c'])
plt.show()
```

```python
sns.distplot(a = df1.mpg )
```

```python
# rug is the jitter : for every entry there is a tick
sns.distplot(a = df1.mpg  , hist = True , kde = True , rug = True)
plt.show()
```

```python
sns.distplot(a = df1.mpg  , hist = True , kde = False , rug = False)
plt.show()
```

```python
sns.distplot(a = df1.mpg  , hist = False , kde = True , rug = False)
plt.show()
```

```python
sns.distplot(a = df1.mpg  , hist = False , kde = False , rug = True)
plt.show()
```

```python
#hist_kws i dict , we pass rwidth to change the parameter
sns.distplot(a = df1.mpg , hist = True , kde = False , rug = False , hist_kws = {"rwidth" : 0.9})
plt.show()
```

```python
sns.distplot(a = df1.mpg , hist = True , kde = False , rug = False , hist_kws = {"rwidth" : 0.9 , "color" : "r"})
plt.show()
```

```python
sns.distplot(a = df1.mpg , hist = False , kde = True , rug = False , kde_kws = {"color" : "r" , "shade" : True})
plt.show()
```

```python
sns.distplot(a = df1.mpg , hist = True , kde = True , rug = False ,hist_kws = {"rwidth" : 0.9 , "color" : "r"}, kde_kws = {"color" : "b" , "shade" : True})
plt.show()
```
