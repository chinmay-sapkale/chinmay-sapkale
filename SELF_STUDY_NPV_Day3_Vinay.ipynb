# Merge

```python
df21 = pd.read_csv("lds21.csv")
df22 = pd.read_csv("lds22.csv")
df23 = pd.read_csv("lds23.csv")

print(df21)
print(df22)
print(df23)
```

```python
pd.merge(left = df21 , right = df22 , on = 'HQ' ,how = 'inner' )
```

```python
pd.merge(left = df21 , right = df22 , on = 'HQ' ,how = 'outer' )
```

```python
#parameter indicator indicates record has been taken from which tables
```

```python
pd.merge(left = df21 , right = df22 , on = 'HQ' ,how = 'inner' ,  indicator = True )
```

```python
pd.merge(left = df21 , right = df22 , on = 'HQ' ,how = 'outer' ,  indicator = True )
```

```python
#suffixes is used to name the columns of left and right table
pd.merge(left = df22 , right = df23 , on = 'HQ' ,how = 'inner' ,  indicator = True ,suffixes = ['_left','_right'] )
```

```python
pd.merge(left = df23 , right = df22 , on = 'HQ' ,how = 'inner' ,  indicator = True ,suffixes = ['_left','_right'] )
```

# Pivot Table

```python
df1 = pd.read_csv("lds1.csv")
df1
```

```python
#index should be categorical data
dp = pd.pivot_table(df1 , index = 'Catg')        #bydefaault aggfunc is mean
print(dp)
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , aggfunc = 'count')      
print(dp)
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , aggfunc = 'max')      
print(dp)
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , aggfunc = 'min')      
print(dp)
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , aggfunc = ['min' ,'max' , np.mean])  # you can we give user define function name like np.mean    
dp
```

```python
dp.columns
```

```python
dp.loc["Pub" ,  ('mean', 'EmpCnt')]
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , values = 'EmpCnt'  , aggfunc = ['min' ,'max' , np.mean])
dp
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , values = ['EmpCnt' , 'Exp']  , aggfunc = ['min' ,'max' , np.mean])
dp
```

```python
df1
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' , columns = 'HQ'  ,values = "EmpCnt" , aggfunc = np.mean)
dp
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' ,columns = 'HQ' , values = 'EmpCnt'  , aggfunc = [np.min ,np.max , np.mean])
dp
```

```python
def mxmndiff(c):
    print(c)
    print(type(c))
    print()
    return c.max() - c.min()
```

```python
dp = pd.pivot_table(df1 , index = 'Catg'  , values = 'EmpCnt'  , aggfunc = mxmndiff)
dp
```

```python
dp = pd.pivot_table(df1 , index = 'Catg' ,columns = 'HQ' , values = 'EmpCnt'  , aggfunc = [np.min , np.max , lambda c : c.max() - c.min()])
dp
```

```python
def vol_class(vol):
    if vol > 950:
        return "Supeb_Sales"
    elif vol < 930:
        return "Average_Sales"
    else:
        return "Good_Sales"
```

```python
df1.Vol.map(vol_class)
```

```python
df1["VolClass"] = df1.Vol.map(vol_class)
df1
```
